```python


df_dataset=df_test[["month","weekday","visit_day","THEFT/LARCENY"]]
display(df_dataset.shape)
display(df_dataset.shape[0])
length=df_dataset.shape[0]
df_train_dataset=df_dataset.iloc[:int(length*0.8),:]
df_test_dataset=df_dataset.iloc[int(length*0.8):,:]
X_train=df_train_dataset[["month","weekday","visit_day"]]
y_train=df_train_dataset[["THEFT/LARCENY"]]
X_test=df_test_dataset[["month","weekday","visit_day"]]
y_test=df_test_dataset[["THEFT/LARCENY"]]

X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
y_train = y_train.fillna(0)
y_test = y_test.fillna(0)
```

```python
#元のデータを破壊しないために、一度df_test_copyにデータを退避。
df_test_copy=df_test.copy()
```

```python
#データをずらす
display(df_test_copy["BURGLARY"].head(5)) #通常
display(df_test_copy["BURGLARY"].shift(1).head(5)) #通常
display(df_test_copy["BURGLARY"].shift(2).head(5)) #通常
display(df_test_copy["BURGLARY"].shift(3).head(5)) #通常


df_test_copy["BURGLARY_before1day"]=df_test_copy["BURGLARY"].shift(1) #新しい"BURGLARY_before1day"のカラムに一日前のデータを追加
```

```python
import lightgbm as lgb



params = {
    'learning_rate': 0.02,
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'rmse',
    'sub_feature': 0.7,
    'num_leaves': 60,
    'min_data': 100,
    'min_hessian': 1,
    'verbose': -1,
}

display(X_train.head(1))
display(X_test.head(1))
display(y_train.head(1))
display(y_test.head(1))

lgb_train = lgb.Dataset(X_train, y_train["THEFT/LARCENY"])
lgb_test = lgb.Dataset(X_test, y_test["THEFT/LARCENY"])

mod = lgb.train(params,
          lgb_train,
          valid_sets=lgb_test,
          num_boost_round=400000,
          early_stopping_rounds=300,
          verbose_eval=1000)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
pred_train = mod.predict(X_train) 
pred_test = mod.predict(X_test)

print("MAE(train):\t%.3f" % (mean_absolute_error(y_train, pred_train)))
print("MAE(test):\t%.3f" % (mean_absolute_error(y_test, pred_test)))
print("RMSE(train):\t%.3f" % (np.sqrt(mean_squared_error(y_train, pred_train))))
print("RMSE(test):\t%.3f" % (np.sqrt(mean_squared_error(y_test, pred_test))))
#print("SMAPE(train):\t%.3f" % (smape(y_train, pred_train)))
#print("SMAPE(test):\t%.3f" % (smape(y_test, pred_test)))

importances = pd.Series(mod.feature_importance(), index = mod.feature_name())
importances = importances.sort_values()
importances.plot(kind = "barh")
plt.title("imporance in the lgbm Model")
plt.show()
```
